```python
days = [1,2,3,4,5,6,7] #days of week 
sales_1 = [160,150,140,145,175,165,180] #sales of comp.1
sales_2 = [70,90,160,150,140,145,175] #sales of comp.2

plt.figure(figsize=(6,5), dpi=150)
plt.title("days vs sales of comp. 1 & 2")
plt.xlabel("days , x axis")
plt.ylabel("sales , y axis")
plt.plot(days, sales_1)
plt.plot(days, sales_2)
```

```python
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,2,1)
plt.title("x vs y_sin")
plt.xlabel("x")
plt.ylabel("y_sin")
plt.plot(x,y_sin,'b--')


plt.subplot(2,2,4)
plt.title("x vs y_cos")
plt.xlabel("x")
plt.ylabel("y_cos")
plt.plot(x,y_cos,'r--')
```
